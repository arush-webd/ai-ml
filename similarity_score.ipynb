```python
import numpy as np
from langchain_huggingface import HuggingFaceEmbeddings

documents=[
    "machine learning is useful",
    "deep learning uses neural networks",
    "stock market prediction using machine learning",
    "I like playing football",
    "natural language processing uses embeddings"
]

query="machine learning for stock preddiction"

embedding_model=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    encode_kwargs={
        "normalize_embeddings":True
    }
)
doc_embeddings=embedding_model.embed_documents(documents)
query_embeddings=embedding_model.embed_query(query)

doc_embeddings=np.array(doc_embeddings)
query_embeddings=np.array(query_embeddings)

def cosine_similarity(a,b):
  return np.dot(a,b)/((np.linalg.norm(a))*(np.linalg.norm(b)))

def similarity_search(doc_embeddings,query_embeddings,docs,k=3):
  scores=[]
  for i in range(len(documents)):
    score=cosine_similarity(query_embeddings,doc_embeddings[i])
    scores.append((score,docs[i]))
  scores.sort(reverse=True,key=lambda x:x[0])
  return scores[:k]

results=similarity_search(
    query_embeddings=query_embeddings,
    doc_embeddings=doc_embeddings,
    docs=documents,
    k=3
)
for score,doc in results:
  print(round(score,4),'->',doc)


```
