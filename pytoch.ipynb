```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define Feedforward Neural Network
class FeedForwardNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(FeedForwardNN, self).__init__()

        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(p=0.5)
        self.fc2 = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.dropout(out)
        out = self.fc2(out)
        return out


# Example usage
input_size = 10
hidden_size = 32
num_classes = 2

model = FeedForwardNN(input_size, hidden_size, num_classes)

# Dummy input
x = torch.randn(5, input_size)

# Forward pass
output = model(x)

print(output)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define CNN
class SimpleCNN(nn.Module):
    def __init__(self, num_classes):
        super(SimpleCNN, self).__init__()

        self.conv1 = nn.Conv2d(
            in_channels=1,
            out_channels=16,
            kernel_size=3,
            stride=1,
            padding=1
        )

        self.relu = nn.ReLU()

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        self.fc = nn.Linear(16 * 14 * 14, num_classes)

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu(x)

        x = self.pool(x)

        # Flatten
        x = x.view(x.size(0), -1)

        x = self.fc(x)

        return x


# Example usage
model = SimpleCNN(num_classes=10)

# Dummy image batch
# Batch size = 8
# Channels = 1
# Image size = 28x28
x = torch.randn(8, 1, 28, 28)

output = model(x)

print(output.shape)
```

```python
import torch

# -----------------------
# Data (y = 2x)
# -----------------------
X = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
y = torch.tensor([[2.0], [4.0], [6.0], [8.0]])

# -----------------------
# Parameters (w, b)
# -----------------------
w = torch.randn(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# -----------------------
# Hyperparameters
# -----------------------
lr = 0.01
epochs = 1000

# -----------------------
# Training loop
# -----------------------
for epoch in range(epochs):

    # Forward pass
    y_pred = w * X + b

    # Loss (MSE)
    loss = ((y_pred - y) ** 2).mean()

    # Backprop
    loss.backward()

    # Gradient descent step
    with torch.no_grad():
        w -= lr * w.grad
        b -= lr * b.grad

    # Reset gradients
    w.grad.zero_()
    b.grad.zero_()

    # Logging
    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch+1}: loss = {loss.item():.4f}")

# -----------------------
# Final parameters
# -----------------------
print("Learned w:", w.item())
print("Learned b:", b.item())
```
