```python
import numpy as np

# -------------------------------------------------------
# 1. ARRAY CREATION
# -------------------------------------------------------

a = np.array([1, 2, 3, 4, 5])
b = np.array([[1., 2., 3.], [4., 5., 6.]])

print("1-D array:", a)
print("2-D array:\n", b)
print("Shape:", b.shape)
print("Dtype:", b.dtype)
print("Ndim:", b.ndim)
print("Size (total elements):", b.size)

# Common initializers
zeros   = np.zeros((3, 4))
ones    = np.ones((3, 4))
eye     = np.eye(3)                  # identity matrix
full    = np.full((2, 3), 7.)
rand    = np.random.randn(3, 4)      # N(0,1)
uniform = np.random.rand(3, 4)       # U(0,1)
arange  = np.arange(0, 10, 2)       # [0 2 4 6 8]
linspace= np.linspace(0, 1, 5)      # [0. .25 .5 .75 1.]

print("\narange:", arange)
print("linspace:", linspace)
```

```python
# -------------------------------------------------------
# 2. INDEXING & SLICING
# -------------------------------------------------------

a = np.array([[10, 20, 30],
              [40, 50, 60],
              [70, 80, 90]])

print("Element [1,2]:", a[1, 2])          # 60
print("Row 0:", a[0])                      # [10 20 30]
print("Col 1:", a[:, 1])                   # [20 50 80]
print("Sub-matrix:\n", a[0:2, 1:3])       # [[20 30][50 60]]
print("Reversed rows:\n", a[::-1])        # flip rows

# Boolean indexing
x = np.array([3, -1, 4, -1, 5, -9])
print("\nPositive only:", x[x > 0])        # [3 4 5]
x[x < 0] = 0
print("Negatives zeroed:", x)

# Fancy indexing (select specific rows/cols)
m = np.random.randint(0, 10, (4, 4))
print("\nMatrix:\n", m)
print("Rows 0 and 2:\n", m[[0, 2]])
print("Diagonal:", m[np.arange(4), np.arange(4)])
```

```python
# -------------------------------------------------------
# 3. RESHAPING & MANIPULATION
# -------------------------------------------------------

a = np.arange(12)
print("Original:", a)

b = a.reshape(3, 4)
print("Reshape (3,4):\n", b)

c = a.reshape(2, 2, 3)
print("Reshape (2,2,3) shape:", c.shape)

# Flatten
print("Flatten:", b.flatten())      # always returns a copy
print("Ravel:", b.ravel())          # returns view when possible

# Transpose
print("Transpose shape:", b.T.shape)

# Add / remove dimensions
x = np.array([1, 2, 3])            # shape (3,)
col = x[:, np.newaxis]              # shape (3,1)  — column vector
row = x[np.newaxis, :]              # shape (1,3)  — row vector
print("\nColumn vector shape:", col.shape)
print("Row vector shape:", row.shape)

# Stack & concatenate
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("\nhstack:", np.hstack([a, b]))          # [1 2 3 4 5 6]
print("vstack:\n", np.vstack([a, b]))          # [[1 2 3][4 5 6]]
print("stack axis=1:\n", np.stack([a, b], axis=1))  # [[1 4][2 5][3 6]]
```

```python
# -------------------------------------------------------
# 4. MATH OPERATIONS & BROADCASTING
# -------------------------------------------------------

a = np.array([[1., 2.], [3., 4.]])
b = np.array([[5., 6.], [7., 8.]])

# Element-wise
print("Add:\n", a + b)
print("Multiply:\n", a * b)         # NOT matrix multiply
print("Matrix multiply:\n", a @ b)  # or np.dot(a, b)

# Scalar broadcast
print("\na * 10:\n", a * 10)
print("a + 100:\n", a + 100)

# Broadcasting rules — shapes are compared right-to-left
# (3,4) + (4,)  -> (3,4)   OK: (4,) broadcasts across rows
# (3,1) + (1,4) -> (3,4)   OK: each dimension stretches

x = np.random.randn(5, 3)       # (5,3)
mean = x.mean(axis=0)           # (3,)
x_centered = x - mean           # (5,3) - (3,) -> (5,3)  broadcast!
print("\nCentered array shape:", x_centered.shape)

# Universal functions (ufuncs)
t = np.array([0., np.pi/2, np.pi])
print("\nSin:", np.sin(t))
print("Exp:", np.exp(np.array([0., 1., 2.])))
print("Log:", np.log(np.array([1., np.e, np.e**2])))
print("Sqrt:", np.sqrt(np.array([1., 4., 9.])))
```

```python
# -------------------------------------------------------
# 5. AGGREGATIONS & STATISTICS
# -------------------------------------------------------

np.random.seed(42)
data = np.random.randn(6, 4)      # 6 samples, 4 features

print("Shape:", data.shape)
print("Mean (all):", data.mean())
print("Mean per column:", data.mean(axis=0))   # shape (4,)
print("Mean per row:", data.mean(axis=1))      # shape (6,)

print("\nStd:", data.std(axis=0))
print("Var:", data.var(axis=0))
print("Min:", data.min(axis=0))
print("Max:", data.max(axis=0))
print("Sum:", data.sum(axis=0))

print("\nArgmin (row index of min per col):", data.argmin(axis=0))
print("Argmax (row index of max per col):", data.argmax(axis=0))

# Cumulative
x = np.array([1, 2, 3, 4, 5])
print("\nCumsum:", np.cumsum(x))
print("Cumprod:", np.cumprod(x))

# Sorting
a = np.array([3, 1, 4, 1, 5, 9, 2])
print("\nSorted:", np.sort(a))
print("Argsort:", np.argsort(a))         # indices that would sort
```

```python
# -------------------------------------------------------
# 6. LINEAR ALGEBRA
# -------------------------------------------------------

A = np.array([[2., 1.],
              [5., 3.]])

# Determinant & inverse
print("Det(A):", np.linalg.det(A))
A_inv = np.linalg.inv(A)
print("Inverse:\n", A_inv)
print("A @ A_inv (should be I):\n", np.round(A @ A_inv, 10))

# Solve linear system  Ax = b
b = np.array([4., 7.])
x = np.linalg.solve(A, b)
print("\nSolution x:", x)
print("Verify Ax:", A @ x)

# Eigenvalues & eigenvectors
vals, vecs = np.linalg.eig(A)
print("\nEigenvalues:", vals)
print("Eigenvectors:\n", vecs)

# SVD
U, S, Vt = np.linalg.svd(A)
print("\nSingular values:", S)

# Norms
v = np.array([3., 4.])
print("\nL2 norm:", np.linalg.norm(v))        # 5.0
print("L1 norm:", np.linalg.norm(v, ord=1))   # 7.0
```

```python
# -------------------------------------------------------
# 7. RANDOM MODULE
# -------------------------------------------------------

rng = np.random.default_rng(seed=0)   # modern API — use this

# Distributions
normal   = rng.normal(loc=0., scale=1., size=(3, 3))
uniform  = rng.uniform(low=0., high=1., size=5)
integers = rng.integers(low=0, high=10, size=5)
binomial = rng.binomial(n=10, p=0.5, size=5)
choice   = rng.choice(np.arange(100), size=10, replace=False)  # sample w/o replacement

print("Normal:\n", normal)
print("Uniform:", uniform)
print("Integers:", integers)
print("Binomial:", binomial)
print("Choice (no replace):", choice)

# Shuffle
arr = np.arange(10)
rng.shuffle(arr)
print("\nShuffled:", arr)

# Reproducibility
rng1 = np.random.default_rng(42)
rng2 = np.random.default_rng(42)
print("\nSame seed same values:", np.allclose(rng1.random(5), rng2.random(5)))
```

```python
# -------------------------------------------------------
# 8. VECTORIZATION vs LOOPS (performance)
# -------------------------------------------------------
import time

n = 1_000_000
a = np.random.randn(n)
b = np.random.randn(n)

# Slow: Python loop
start = time.time()
c_loop = [a[i] * b[i] for i in range(n)]
print(f"Loop:   {time.time() - start:.4f}s")

# Fast: vectorized
start = time.time()
c_vec = a * b
print(f"Numpy:  {time.time() - start:.4f}s")

# np.vectorize — wraps a scalar function for array input
# (convenience, NOT truly fast — still loops under the hood)
def clamp(x, lo, hi):
    return max(lo, min(hi, x))

vclamp = np.vectorize(clamp)
x = np.array([-2., 0.5, 3., 1.2])
print("\nClamped [0,1]:", vclamp(x, 0, 1))

# np.where — vectorized conditional
result = np.where(x > 0, x, 0.)    # ReLU
print("ReLU via np.where:", result)

# Avoid copies with in-place ops
x = np.ones(5)
x += 1          # in-place, no extra allocation
x *= 2
print("\nIn-place result:", x)
```

```python
# -------------------------------------------------------
# 9. PRACTICAL: NORMALIZATION, CORRELATION, MISSING DATA
# -------------------------------------------------------

np.random.seed(7)
X = np.random.randn(100, 4)   # 100 samples, 4 features

# --- Min-Max normalization ---
X_min = X.min(axis=0)
X_max = X.max(axis=0)
X_minmax = (X - X_min) / (X_max - X_min)
print("Min-max range:", X_minmax.min(axis=0).round(2),
      "to", X_minmax.max(axis=0).round(2))

# --- Z-score standardization ---
X_mean = X.mean(axis=0)
X_std  = X.std(axis=0)
X_std_norm = (X - X_mean) / X_std
print("Z-score mean:", X_std_norm.mean(axis=0).round(6))
print("Z-score std: ", X_std_norm.std(axis=0).round(6))

# --- Correlation matrix ---
corr = np.corrcoef(X.T)     # shape (4, 4)
print("\nCorrelation matrix:\n", corr.round(2))

# --- Handling NaNs ---
data = np.array([1., 2., np.nan, 4., np.nan, 6.])
print("\nHas NaN:", np.isnan(data).any())
print("NaN count:", np.isnan(data).sum())
print("Mean (ignoring NaN):", np.nanmean(data))
print("Std  (ignoring NaN):", np.nanstd(data))

# Replace NaN with column mean
col_mean = np.nanmean(data)
data[np.isnan(data)] = col_mean
print("After fill:", data)
```
